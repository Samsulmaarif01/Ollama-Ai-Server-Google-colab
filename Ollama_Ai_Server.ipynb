```python
from google.colab import userdata
from pyngrok import ngrok
import time
import subprocess
import threading
import os

# Environment agar Ollama bisa diakses dari luar
os.environ["OLLAMA_HOST"] = "0.0.0.0:11434"
os.environ["OLLAMA_ORIGINS"] = "*"

# Ambil token ngrok dari secret (ganti 'NGROK_AUTH_TOKEN' dengan nama secret Anda atau authToken langsung)
NGROK_AUTH_TOKEN = userdata.get('NGROK_AUTH_TOKEN')
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# Jalankan server Ollama di background
def run_ollama():
    subprocess.Popen(["ollama", "serve"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

thread = threading.Thread(target=run_ollama)
thread.start()
time.sleep(5)  # Tunggu server siap

# Buat tunnel untuk port 11434 (Ollama API)
public_url_api = ngrok.connect(11434, "http").public_url
print(f"\n🌐 URL PUBLIK API OLLAMA (untuk OpenCode CLI):\n{public_url_api}/v1\n")
print("Simpan URL ini untuk konfigurasi opencode.json Anda.")
```
